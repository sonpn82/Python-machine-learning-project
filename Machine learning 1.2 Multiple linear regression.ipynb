```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sonpn82/Python-machine-learning-project/master/FuelConsumptionCo2.csv')
df.head()
```

```python
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(9)
```

```python
# plot emission value vs enginer size
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

## Split train and test data set

```python
msk = np.random.rand(len(df)) < 0.8  # create a len(df) random numbers from 0-1, compare with 0.8 to get True or False output
train = cdf[msk]
test = cdf[~msk]
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue')
plt.xlabel("Engine size")
plt.ylabel("CO2 emission")
plt.show()
```

## Multiple regression model

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit(x, y)

# print coefficients
print('Coefficients: ', regr.coef_)
```

Ordinary least square is used for estimate coefficients of above equation <br/>
Direct solver or optimization algorithm (gradient descent, stochastic gradient descent...) can be used to optain best parameters

## Prediction

```python
y_hat= regr.predict(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
x = np.asanyarray(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(test[['CO2EMISSIONS']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
